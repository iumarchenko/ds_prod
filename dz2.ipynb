### Напишите функцию, которая возвращает сумму двух вещественных аргументов a и b, а к ней декоратор, который делает так, чтобы возвращаемое значение функцией было по модулю 5.


```python
def mod_5(func):
    def wrapper(*args, **kwargs):
        return_value = func(*args, **kwargs)
        return return_value % 5
    return wrapper

@mod_5
def my_func(a,b):
    return a+b

dummy = my_func(3,5)
print(dummy)
```

### Попробуйте описать своими словами основные отличия SMOTE от ADASYN, ознакомившись с документацией к ним.

Алгоритм SMOTE - метод увеличения числа примеров миноритарного класса (Synthetic Minority Oversampling Technique)

Этот метод генерирует дополнительные данные для увеличения экземпляров меньшего класса на основе сходства пространств объектов между существующими экземплярами меньшинства. Новые элементы похожи на уже имеющиеся, но не дублируют их. <br>
Чтобы создать новый экземпляр, алгоритм находит K-ближайших соседей каждого экземпляра меньшинства, случайным образом выбирает один из них, а затем создает новые экземпляры путем случайного выбора точки на линии, соединяющей выбранный экземпляр с ближайшим соседом

ADASYN - Адаптивный синтез (ADASYN) 

Алгоритм генерирует экземпляры класса меньшинства в соответствии с их распределением плотности. <br>
Он вычисляет K-ближайших соседей каждого экземпляра меньшинства, а затем получает соотношение классов экземпляров меньшинства и большинства для создания новых выборок. <br>
Для образцов, принадлежащих к меньшинству, у которых рядом много экземпляров класса большинства, генерируется больше новых данных по сравнению с образцами меньшинства, у которых рядом меньше классов большинства. <br>
Генерация может быть с использованием линейной комбинации выбранного экземпляра и соседа (аналогично SMOTE), или с использованием более продвинутых методов, таких как рисование плоскости между тремя наблюдениями меньшинства и случайный выбор точки на этой плоскости.
